```python
from hailo_sdk_client import ClientRunner

model_name = 'mnist'
quantized_model_har_path = '{}_quantized_model.har'.format(model_name)
runner = ClientRunner(hw_arch='hailo8', har_path=quantized_model_har_path)
```

```python
# compiles the model returning a binary that contains the compiled model, a HEF file.
hef = runner.compile()

#writes binary in a file
file_name = model_name + '.hef'
with open(file_name, 'wb') as f:
    f.write(hef)
```
